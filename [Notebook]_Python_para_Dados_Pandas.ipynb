```python
df_concurso = pd.read_csv('dados_concurso.csv')
df_concurso.head()
```

```python
from datetime import datetime, date
```

```python
def age(born): #calcula idade
    if pd.isna(born):  # Check for missing values
        return None
    born = str(born)
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month,
                                      today.day) < (born.month,
                                                    born.day))
```

```python
age = df_concurso['Data de Nascimento'].apply(age) #gera coluna de idades
age
```

```python
df_concurso['Idade'] = age #adiciona colunas de idade
df_concurso.head()
```
